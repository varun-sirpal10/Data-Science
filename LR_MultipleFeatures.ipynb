```python
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X.shape)
```

```python
def hypothesis(x,theta):
    y_ = 0.0
    n = X.shape[1]
    for i in range(n):
        y_ += theta[i]*x[i]
    return y_
```

```python
def error(X,Y,theta):
    m = X.shape[0]
    error = 0.0
    for i in range(m):
        y_ = hypothesis(X[i],theta)
        error += (y_ - Y[i])**2
    return error/m
```

```python
def gradient(X,Y,theta):
    m,n = X.shape
    grad = np.zeros((n,))
    
    for j in range(n):
        for i in range(m):
            y_ = hypothesis(X[i],theta)
            grad[j] += (y_ - Y[i])*X[i][j]
    return grad/m
```

```python
def gradient_descent(X,Y,learning_rate=0.1,max_epochs=300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epochs):
        err = error(X,Y,theta)
        error_list.append(err)
        grad = gradient(X,Y,theta)
        
        for j in range(n):
                theta[j] = theta[j] - learning_rate*grad[j]
    
    return theta,error_list
```

```python
theta,error_list = gradient_descent(X,Y)
```

```python
plt.plot(error_list)
plt.show()
```

## Efficient Code using Vectorisation

```python
def hypothesis(X,theta):
    return np.dot(X,theta)
```

```python
def error(X,Y,theta):
    e = 0.0
    y_ = hypothesis(X,theta)
    e = np.sum((y_ - Y)**2)
    
    return e/m
```

```python
def gradient(X,Y,theta):
    m = X.shape[0]
    y_ = hypothesis(X,theta)
    grad = np.dot(X.T,(y_ - Y))
    
    return grad/m
```

```python
def gradient_descent(X,Y,learning_rate=0.1,max_epochs=300):
    m,n = X.shape
    theta = np.zeros((n,))
    error_list = []
    
    for i in range(max_epochs):
        err = error(X,Y,theta)
        error_list.append(err)
        grad = gradient(X,Y,theta)
        
        theta = theta - learning_rate*grad
    
    return theta,error_list
```

```python
theta,error_list = gradient_descent(X,Y)
```
