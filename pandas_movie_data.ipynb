## Pandas - Working with Movie Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("movie_metadata.csv")
df.head(n=10)
```

```python
df.columns
```

```python
score = df.get('imdb_score')
```

```python
score
```

```python
titles = df.get('movie_title')
#print(titles)
```

```python
freq_title = {}

for title in titles:
    length = len(title)
    
    try:
        freq_title[length] += 1
    except:
        freq_title[length] = 1
```

```python
X = list(freq_title.keys())
Y = list(freq_title.values())
```

```python
plt.scatter(X,Y)
plt.show()
```
